```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
#Read stats of each Asian country
train0 = pd.read_csv('Downloads/2022WCKOREA.csv') #South Korea
train1 = pd.read_csv('Downloads/2022WCJAPAN.csv') #Japan
train2 = pd.read_csv('Downloads/2022WCAUSTRALIA.csv') #Australia
train3 = pd.read_csv('Downloads/2022WCIRAN.csv') #Iran
train4 = pd.read_csv('Downloads/2022WCSAUDIARABIA.csv') #SaudiArabia
train5 = pd.read_csv('Downloads/2022WCQATAR.csv') #Qatar
```

```python
#Train data of South Korea
display(train0[0:2].T)
```

```python
#Print the name of columns
cols=train0.columns.tolist()
print(cols)
print(len(cols)) #number of columns
```

```python
for i in range(7,29):
 data = train0.sort_values(cols[i],ascending=False)[0:40]
#sort 'npxG+xAG values' in ascending order
plt.figure(figsize=(15,4))
plt.title(cols[i])
g = sns.barplot(x=data['Player'],y=data[cols[i]],color="red")
g.tick_params(axis='x', labelrotation=90)
plt.show()
```
